```python
import pandas as pd
import matplotlib.pyplot as plt

# Load log
df = pd.read_csv("training_log.csv")

models = df['model'].unique()

# Plot Loss
plt.figure()
for m in models:
    sub = df[df['model'] == m]
    plt.plot(sub['epoch'], sub['loss'], label=m)

plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training Loss")
plt.legend()
plt.grid()
plt.show()

# Plot Accuracy
plt.figure()
for m in models:
    sub = df[df['model'] == m]
    plt.plot(sub['epoch'], sub['val_acc'], label=m)

plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.title("Training Accuracy")
plt.legend()
plt.grid()
plt.show()

# Plot Macro F1
plt.figure()
for m in models:
    sub = df[df['model'] == m]
    plt.plot(sub['epoch'], sub['val_f1'], label=m)

plt.xlabel("Epoch")
plt.ylabel("Macro F1")
plt.title("Training Macro F1")
plt.legend()
plt.grid()
plt.show()
```

```python
# Lấy best F1 của mỗi model
best_f1 = df.groupby("model")["val_f1"].max().reset_index()

plt.figure()
plt.bar(best_f1["model"], best_f1["val_f1"])

plt.xlabel("Model")
plt.ylabel("Best Macro F1")
plt.title("Best Macro F1 Comparison")
plt.xticks(rotation=30)
plt.grid(axis='y')

# Hiển thị số trên cột
for i, v in enumerate(best_f1["val_f1"]):
    plt.text(i, v + 0.5, f"{v:.2f}", ha='center')

plt.show()
```
